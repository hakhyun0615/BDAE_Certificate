# Log 변환

```python
import pandas as pd
data = pd.read_csv('Ex_CEOSalary.csv',encoding='utf-8')
del data['industry']
```

```python
import numpy as np
data['log_salary'] = np.log(data['salary'])
data['log_sales'] = np.log(data['sales'])
data['log_roe'] = np.log(data['roe'])
data.head()
```

```python
data.hist(bins=50,figsize=(20,15)) # 큰 값들에 의한 왜곡된 관계가 해결되어 데이터가 중앙으로 이동한 효과가 있었음
```

```python
data.corr()
```

# 제곱근 변환

```python
data['sqrt_salary'] = np.sqrt(data['salary'])
data['sqrt_sales'] = np.sqrt(data['sales'])
data['sqrt_roe'] = np.sqrt(data['roe'])
data.head()
```

```python
data.hist(bins=15,figsize=(20,15)) # 로그변환과 유사하게 정규분포에 근접해졌다.
```

```python
data.corr()
```
