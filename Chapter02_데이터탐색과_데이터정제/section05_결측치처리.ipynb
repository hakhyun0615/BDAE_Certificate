# 결측값 대체: fillna(바꾸고자 하는 값이나 문자)

```python
data = pd.read_csv('Ex_Missing.csv')
data
```

```python
data_0 = data.fillna(0)
data_0
```

```python
data_missing = data.fillna('missing')
data_missing
```

```python
# 결측값을 해당 변수의 앞의 값으로 대체
data_ffill = data.fillna(method='ffill')
data_ffill
```

```python
# 결측값을 해당 변수의 다음 값으로 대체
data_bfill = data.fillna(method='bfill')
data_bfill
```

```python
# 각 변수의 평균으로 대체
data_mean = data.fillna(data.mean())
data_mean
```

```python
# 각 변수의 중위수로 대체
data_median = data.fillna(data.median())
data_median
```

```python
# 각 변수의 최대값으로 대체
data_max = data.fillna(data.max())
data_max
```

```python
# 다른 변수의 평균으로 대체
data_other_mean = data.fillna(data['salary'].mean())
data_other_mean
```

```python
# 다른 변수의 값으로 대체
import numpy as np
data['sales_new'] = np.where(pd.notnull(data['sales'])==True,data['sales'],data['salary'])
data
```

```python
# 집단별 평균값으로 대체
fill_mean_func = lambda g: g.fillna(g.mean())
data_group_mean = data.groupby('industry').apply(fill_mean_func)
data_group_mean
```

```python
# 집단별 사용자 지정 특정 값으로 대체
fill_values = {1:1000,2:2000}
fill_func = lambda d: d.fillna(fill_values[d.name])
data_group_value = data.groupby('industry').apply(fill_func)
data_group_value
```

```python
# 변수별 다른 대체방법
missing_fill_val = {'salary':data.salary.interpolate(), 'sales':data.sales.mean(), 'roe':'missing'}
data_multi = data.fillna(missing_fill_val)
data_multi
```
