```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import GradientBoostingRegressor

# read
data = pd.read_csv('house_price.csv')

# X/y
X = data[['housing_age', 'income','bedrooms', 'households', 'rooms']]
y = data[['house_value']]

# train/test
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=42)

# X cat dummies,align

# X num 정규화
scaler = MinMaxScaler()
scaler.fit(X_train)
X_scaled_train = scaler.transform(X_train)
X_scaled_test = scaler.transform(X_test)

# X final
X_final_train = X_scaled_train
X_final_test = X_scaled_test

# y
y_train = y_train['house_value']

# 모델 학습
model = GradientBoostingRegressor()
model.fit(X_final_train, y_train)

# 모델 학습 결과
pred_test = model.predict(X_final_test)
y_test['y_pred'] = pred_test
final_result = pd.concat([X_test,y_test],axis=1)
final_result
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test,pred_test)
np.sqrt(mse)
```
