```python
# 7(1+3+1+2)
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression

# read
X_train = pd.read_csv('X_train.csv')
X_test = pd.read_csv('X_test.csv')
y_train = pd.read_csv('y_train.csv')

# X cat dummies,align
X_train_num = X_train[['Customer_care_calls', 'Cost_of_the_Product', 'Prior_purchases', 'Discount_offered', 'Weight_in_gms']]
X_train_cat = pd.get_dummies(X_train[['Warehouse_block', 'Mode_of_Shipment', 'Customer_rating', 'Product_importance','Gender']])
X_test_num = X_test[['Customer_care_calls', 'Cost_of_the_Product', 'Prior_purchases', 'Discount_offered', 'Weight_in_gms']]
X_test_cat = pd.get_dummies(X_test[['Warehouse_block', 'Mode_of_Shipment', 'Customer_rating', 'Product_importance','Gender']])
X_train_cat, X_test_cat = X_train_cat.align(X_test_cat, join='inner', axis=1)

# X num 정규화
scaler = MinMaxScaler()
scaler.fit(X_train_num)
X_scaled_train_num  = scaler.transform(X_train_num)
X_scaled_test_num = scaler.transform(X_test_num)

# X final
X_train_final = pd.concat([pd.DataFrame(X_scaled_train_num), X_train_cat], axis=1)
X_test_final = pd.concat([pd.DataFrame(X_scaled_test_num), X_test_cat], axis=1)

# y
y_train = y_train['Reached.on.Time_Y.N']

# 모델 학습
model = LogisticRegression()
model.fit(X_train_final,y_train)

# 모델 학습 결과
pred_test = model.predict(X_test_final)
pred_proba_test = model.predict_proba(X_test_final)
pred_proba_test = pd.DataFrame(pred_proba_test[:,0],columns=['predict_proba'])
pred_final = pd.concat([X_test['ID'],pred_proba_test], axis=1)
pred_final.to_csv('predict_practice.csv', index=False)
```

```python
# 1 + 3 + 1 + 2 (7)
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression

# read
test = pd.read_csv('X_test.csv')
X = pd.read_csv('X_train.csv')
y = pd.read_csv('y_train.csv')

# X의 cat dummies,align
X_num = X[[ 'Customer_care_calls','Customer_rating', 'Cost_of_the_Product', 'Prior_purchases', 'Discount_offered', 'Weight_in_gms']]
X_cat = pd.get_dummies(X[['Warehouse_block', 'Mode_of_Shipment','Product_importance', 'Gender']])
test_num = test[[ 'Customer_care_calls','Customer_rating', 'Cost_of_the_Product', 'Prior_purchases', 'Discount_offered', 'Weight_in_gms']]
test_cat = pd.get_dummies(test[['Warehouse_block', 'Mode_of_Shipment','Product_importance', 'Gender']])
X_cat, test_cat = X_cat.align(test_cat, join='inner', axis=1)

# X의 num 정규화
minmax = MinMaxScaler()
minmax.fit(X_num)
X_scaled = minmax.transform(X_num)
test_scaled = minmax.transform(test_num)

# X의 final
X_final = pd.concat([pd.DataFrame(X_scaled),X_cat], axis=1)
test_final = pd.concat([pd.DataFrame(test_scaled),test_cat], axis=1)

# y
y = y['Reached.on.Time_Y.N']

# 모델 학습
model = LogisticRegression()
model.fit(X_final, y)

# 모델 학습 결과
pred_test = model.predict_proba(test_final)
pred_test_prob = pd.DataFrame(pred_test[:, 0], columns=['predict_prob'])
final_predict = pd.concat([test['ID'], pred_test_prob], axis=1)
final_predict.to_csv('predict.csv', index=False)
```
